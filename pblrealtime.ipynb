```python
import cv2
from ultralytics import YOLO
from matplotlib import pyplot as plt
import time

# Load the YOLOv8 model
model = YOLO("C:/Users/Rutuja Navale/Desktop/PBLFINAL/my_model/my_model.pt")

# Start the webcam
cap = cv2.VideoCapture(0)

# Set frame dimensions (optional)
cap.set(3, 640)
cap.set(4, 480)

frame_count = 0
max_frames = 10  # Change this if you want to show more/less frames

while frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Inference
    results = model(frame)

    # Annotate frame
    annotated_frame = results[0].plot()

    # Display using matplotlib (suitable for Jupyter)
    plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    frame_count += 1
    time.sleep(0.5)  # Add a small delay between frames

# Release resources
cap.release()
```
